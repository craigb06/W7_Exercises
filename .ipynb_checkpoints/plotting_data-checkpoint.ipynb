# Plotting Data

```python
# Import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
# Create x and y variable then plot the line graph for x and y
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.show()
```

```python
# Update previous code cell to add markers to data points
plt.plot(x, y, marker='o')
plt.show()
```

```python
# Update previous code cell to add title and labels
plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Styling Plots

```python
# Update previous code cell to dashed line style and line color red
plt.plot(x, y, 'o--r')
plt.show()
```

```python
# Update previous code cell to change marker style to square, size to 10, and color to green
plt.plot(x, y, marker='s', ms=10, mfc='g')
plt.show()
```

```python
# Experiment with at least two different marker styles and colors
# Display the results
plt.plot(x, y, marker='*', ms=25, mfc='k')
plt.show()
plt.plot(x, y, marker='>', ms=7, mfc='y')
plt.show()
```
